# Making Some Predictions

```python
predictions = []
for idx, row in test_data.iterrows():
    if row['Sex'] == 'female': 
        if row['Pclass'] == 1:
            predictions.append(1)
        elif row['Pclass'] == 2:
            predictions.append(1)
        elif row['Pclass'] == 3:
            if row['Age'] <= 7:
                predictions.append(1)
            elif row['Age'] > 7:
                    if row['Embarked'] == 'C':
                        predictions.append(1)
                    else:
                        predictions.append(0)
            else:
                predictions.append(0)
        else:
            predictions.append(0)
    elif row['Sex'] == 'male':
        if row['Pclass'] == 1:
            if row['Age'] <=17:
                predictions.append(1)
            else:
                predictions.append(0)
        elif row['Pclass'] == 2:
            if row['Age'] <= 15:
                predictions.append(1)
            else:
                predictions.append(0)
        else:
            predictions.append(0)
            
    else:
        predictions.append(0) # perished
```

```python
len(predictions)
```

```python
test_data['Survived'] = predictions
```
