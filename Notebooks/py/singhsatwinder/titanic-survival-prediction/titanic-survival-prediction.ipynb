```python
data=pd.read_csv("../input/train.csv")
```

```python
data.head()
```

```python
data['Cabin'].isnull().value_counts()
```

```python
data.drop(['Name','Ticket','Cabin'],inplace=True,axis=1)
```

```python
data['Age'].isnull().value_counts()
```

```python
import seaborn as sns
sns.heatmap(data.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python

sns.barplot(x='Pclass',y='Age',data=data)
```

```python
def age_fill(cols):
    Age=cols[0]
    Pclass=cols[1]
    
    if pd.isnull(Age):
        
        if Pclass==1:
            return 37
        elif Pclass==2:
            return 29
        else:
            return 24
    else:
        return Age
```

```python
data['Age']=data[['Age','Pclass']].apply(age_fill,axis=1)
```

```python
data.head(10)
```

```python
data['Sex']=data['Sex'].apply(lambda x : 1 if x=='male' else 0 )
```

```python
Embarked=pd.get_dummies(data['Embarked'],drop_first=True)
```

```python
data.drop('Embarked',axis=1,inplace=True)
```

```python
data=pd.concat([data,Embarked],axis=1)
```

```python
data.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression() 
```

```python
X=data.drop('Survived',axis=1)
y=data['Survived']
```

```python
model.fit(X,y)
```
