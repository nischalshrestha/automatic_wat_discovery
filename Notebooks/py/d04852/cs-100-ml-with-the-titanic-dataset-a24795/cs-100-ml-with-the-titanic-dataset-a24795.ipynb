# Making Some Predictions

```python
predictions = []
for idx, row in test_data.iterrows():
        if row['Sex'] == 'female' and row['SibSp'] < 2:
            predictions.append(1)
        else:
            if row['Age'] < 10:
                predictions.append(1)
            else:
                predictions.append(0)
```

```python
len(predictions)
```

```python
test_data['Survived'] = predictions
```
