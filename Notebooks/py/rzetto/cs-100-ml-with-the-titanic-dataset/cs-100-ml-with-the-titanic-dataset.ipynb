# Working with Rows

# Making Some Predictions

```python
predictions = []
for idx, row in test_data.iterrows():
    if row['Sex'] == 'female':
        if row['Pclass'] < 3:
            predictions.append(1) # survived
        else:
            if(row['Fare'] < 20):
                # age<10 u lived, >10 was 50%, noAge 7:5
                predictions.append(1)
            else:
                predictions.append(0)
    else:
        if (row['Age'] < 10):
            predictions.append(1)
        else:
            predictions.append(0)
```

```python
len(predictions)
```

```python
test_data['Survived'] = predictions
```
