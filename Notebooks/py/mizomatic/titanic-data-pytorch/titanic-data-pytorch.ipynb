```python
class Bin_Classifier(nn.Module):
    def __init__(self):
        super(Bin_Classifier,self).__init__()
        self.input = nn.Linear(6,64)
        self.output = nn.Linear(64,2)
        
    def forward(self,x):
        x = self.input(x)
        x = self.output(x)
        return(nn.Sigmoid(x))
```
