```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
def calculate_age(cols):
    Age=cols[0]
    Pclass=cols[1]
    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass==2:
            return 29
        else:
            return 24
    else:
        return Age
```

```python
train[['Age','Pclass']].head()
```

```python
train['Age']=train[['Age','Pclass']].apply(calculate_age,axis=1)
```

```python
sns.heatmap(train.isnull(),cmap='viridis',yticklabels=False,cbar=False)
```

```python
train.drop('Cabin',axis=1,inplace=True)
```

```python
sns.heatmap(train.isnull(),cmap='viridis',yticklabels=False,cbar=False)
```

```python
train.dropna(inplace=True)
```

```python
sns.heatmap(train.isnull(),cmap='viridis',yticklabels=False,cbar=False)
```

```python
train.head()
```

```python
pd.get_dummies(train['Sex']).head(2)
```

```python
sex=pd.get_dummies(train['Sex'],drop_first=True)
```

```python
embark=pd.get_dummies(train['Embarked'],drop_first=True)
```

```python
train=pd.concat([train,sex,embark],axis=1)
```

```python
train.head(2)
```

```python
train.drop('PassengerId',axis=1,inplace=True)
```

```python
train.head()
```

```python
PC=pd.get_dummies(train['Pclass'],drop_first=True)
```

```python
train=pd.concat([train,PC],axis=1)
```

```python
train.head()
```

```python
#train.drop([2,3],axis=1,inplace=True)
```

```python
train.drop(['Embarked'],axis=1,inplace=True)
```

```python
train.drop(['Name','Sex','Ticket','Pclass'],axis=1,inplace=True)
```

```python
X=train.drop('Survived',axis=1)
y=train['Survived']
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression()
```

```python
logmodel.fit(X_train,y_train)
```

```python
predictions=logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,predictions)
```
