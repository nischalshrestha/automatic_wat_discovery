```python
# To support both python 2 and python 3
from __future__ import division, print_function, unicode_literals

# To plot pretty figures
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
from scipy.stats import mode
import string
import os

train = pd.read_csv("../input/train.csv")
test = pd.read_csv("../input/test.csv")

train_test = [train, test]
train.shape
```

```python
train.head()
```

```python
train.describe()
```

```python
cols = ['Name','Ticket','Cabin']
train = train.drop(cols,axis=1)
```

```python
train.head()
```

```python
train = train.dropna()
```

```python
ds = []
cols = ['Pclass','Sex','Embarked']
for col in cols:
 ds.append(pd.get_dummies(train[col]))
```

```python
titanic_ds = pd.concat(ds, axis=1)
```

```python
train = pd.concat((train,titanic_ds),axis=1)
```

```python
train = train.drop(['Pclass','Sex','Embarked'],axis=1)
```

```python
train['Age'].fillna(train['Age'].median(), inplace=True)
```

```python
X = train.values
y = train['Survived'].values
```

```python
X = np.delete(X,1,axis=1)
```

```python
from sklearn import cross_validation
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X,y,test_size=0.35,random_state=0)
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier(max_depth=5)
clf.fit(X_train,y_train)
clf.score(X_test,y_test)
```
